```python
# from this, we just want to extract the list of 'values of name key' -> ['Action', 'Adventure', 'Science Fiction']
# but the problem is 'genres' is not exacly list, instead it is in the form of 'string of list', so we have to convert it into list using 'ast.literal_eval()' function
```

```python
import ast
```

```python
# this func will first convert the 'striing of list' into string of list, then extract the list of genres
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
def convertCrew(obj):
    L = []
    for i in ast.literal_eval(obj):
        if(i['job'] == 'Director'):
            L.append(i['name'])
            break
    return L
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python

def stem(text):
    L = []
    #first we will convert string into list of words
    for i in text.split():
        #now we will stem every word to it's root word
        L.append(ps.stem(i))
    #now we will convert these words into sring and return them
    return " ".join(L)
```

```python
new_df['tags'].apply(stem)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

## Now let's do text Vectorization

```python

# we will create vectors of all the movies, and if a user like a movie and want the similar suggestions, we will recommend the closest 5 vectors(movies)
# for this, we are going to use the technique of bag of words(we can also use tf-IDF, word2vector)
# In bag of words we count the frequency of words and take out maybe 5000(you can take any number, but less words=more functionality and small table) most repeated words across all the tags of movies, and then we will make table between 5000 words and 5000 movies, and every movie will convert into vector with all it's tags, and which tags are repeated the most across 5 movies which get suggested
# all the stop words will first be removed
# we will use 'scikit learn' to perform all thsi task, using function 'count vectorizer'
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
cv.fit_transform(new_df['tags']).toarray()
# we are converting the resultant table into array as it will contain so many "0's"
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

## Let's calculate the distance(similarity) between Movies

```python
## Now we have to calculate the distance between movies(vectors), based on cosine distance(angular measure) instead of eucleadian distance
# for data with large distance eucledian is not really reliable, so we are using cosine(Î¸),
# so movies with less cosine distance can be suggested, as less distance = more similarity
# so we are going to import 'cosine similarity' function from sklearn
```

```python
from sklearn.metrics.pairwise import cosine_similarity
#if the result is closer to '1'-> 'More similarity', and if result is closer to '0' -> 'less similarity'
```

```python
cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
similarity = cosine_similarity(vectors)
```

## Let's make the Main Recommendation function

```python
# this function will take the given movie, and recommend 5 similar movies
# we will first find the index of the movie, then with this index we will find it in similarity score array, and sort the scores and fetch first 5 movies
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    #after sorting, we will lose that what distance is for what movie, for that we can use enumerate function, it will keep track of distance for all movies
    #and if we use sorting as it is, it will sort on the basis of index, so we would have to give it the index of the score to sort on->using lambda
    #and then we will extract the first 5 movies [1:6]
    movies_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x: x[1])[1:6]
    for i in movies_list:
        #print(i[0]) #this will print only index of movies, not the similarity score as well
        ##but we have nothing to do with these indices, we need names of the movies, so we will use these indices to fetch the movie name
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('Batman Begins')
```
